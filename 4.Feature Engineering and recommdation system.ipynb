```python
#  import TfidfVectorizer, and cosine_similarity. 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Merge answers with professionals to get professional details
answers_prof = answers.merge(professionals, on='author_id', how='left')
answers_prof.head()
```

```python
# Combine question titles and bodies for TF-IDF vectorization
data_merged['questions_content'] = data_merged['questions_title'] + " " + data_merged['questions_body']
```

```python
data_merged.rename(columns={'author_id_x':'author_id','num_questions_answered_x':'num_questions_answered'}, inplace=True)
data_merged.drop(columns=['author_id_y','num_questions_answered_y'], inplace=True)
```

```python
# Content Similarity
def compute_similarity(text1, text2):
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform([text1, text2])
    similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    return similarity[0][0]
```

```python
data_merged['content_similarity'] = data_merged.apply(lambda x: compute_similarity(x['questions_content'],x['answers_body']),axis=1)
```

```python
# Aggregate the duplicate entries by taking the mean of content_similarity
data_aggregated = data_merged.groupby(['author_id', 'question_id'], as_index=False).agg({'content_similarity': 'mean'})
```

```python
user_item_matrix = data_aggregated.pivot(index='author_id', columns='question_id', values='content_similarity').fillna(0)
```

```python
from sklearn.decomposition import TruncatedSVD
# Number of latent factors
n_latent_factors = 10

# Create the SVD model
svd = TruncatedSVD(n_components=n_latent_factors)
```

```python
# Fit the model to the user-item matrix
U = svd.fit_transform(user_item_matrix)
Sigma = svd.singular_values_
Vt = svd.components_
```

```python
import numpy as np
# Reconstruct the matrix
reconstructed_matrix = np.dot(np.dot(U, np.diag(Sigma)), Vt)
```

```python
# Convert the reconstructed matrix back to a DataFrame
reconstructed_df = pd.DataFrame(reconstructed_matrix, index=user_item_matrix.index, columns=user_item_matrix.columns)
```

```python
def recommend_questions(professional_id, original_df, reconstructed_df, num_recommendations=5):
    # Get the professional's row in the reconstructed matrix
    professional_row = reconstructed_df.loc[professional_id]
    
    # Sort the questions by predicted interaction score
    recommended_questions = professional_row.sort_values(ascending=False)
    
    # Filter out already interacted questions
    interacted_questions = original_df[original_df['author_id'] == professional_id]['question_id'].values
    recommended_questions = recommended_questions.drop(labels=interacted_questions)
    
    # Return top N recommendations
    return recommended_questions.head(num_recommendations).index.tolist()
```

```python
# Example: Get recommendations for a professional with ID 'prof_1'
recommendations = recommend_questions('cbd8f30613a849bf918aed5c010340be', data_aggregated, reconstructed_df)
print("Recommended questions:", recommendations)
```

```python
# Create a dictionary to store recommendations for each professional
recommendations_dict = {}
# Iterate over each unique professional ID
for author_id in data_aggregated['author_id'].unique():
    try:
        recommendations = recommend_questions(author_id, data_aggregated, reconstructed_df)
        recommendations_dict[author_id] = recommendations
    except KeyError:
        # Handle the case where the professional ID is not found in the reconstructed_df
        print(f"Author_id ID {author_id} not found in the reconstructed DataFrame.")
```

```python
# Convert the recommendations dictionary to a DataFrame
recommendations_df = pd.DataFrame.from_dict(recommendations_dict, orient='index')
recommendations_df.columns = [f"Recommendation_{i+1}" for i in range(recommendations_df.shape[1])]
```
